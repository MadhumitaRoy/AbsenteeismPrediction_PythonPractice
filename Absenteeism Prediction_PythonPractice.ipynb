# Feature Engineering

Reason for Absence

It is seen from the feature description of the reason for absence that reasons from 1-14 are related to some disease, reasons from 15-17 are related to pregnancy, reasons from 18 to 21 are related to poisoning and reasons from 21-28 are others. Therefore they can be grouped into broader categories. This is mainly because we have only 700 data. Having 4 categories instead of 28 categories for 700 data will provide more insights.

```python
# GRouping reason for absence into 4 major categories

def reason_type(column):
    reason = column[0]
    
    if reason <= 14:
        return 1
    elif (15 <= reason <= 17):
        return 2
    elif (18 <= reason <= 21):
        return 3
    else:
        return 4
```

```python
df["Reason for Absence"] = df[["Reason for Absence"]].apply(reason_type,axis=1)
```

# Standardizing the X features excluding the dummy variables

Note: The dummy variables are not standardized (mean = 0 and SD = 1) as they are already binary.

```python
Unscaled_X = df_preprocessed.drop(['Reason for Absence_2', 'Reason for Absence_3', 'Reason for Absence_4',
       'Education_1','Absenteeism Time in Hours'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
Scaled_X = scaler.fit_transform(Unscaled_X)
Scaled_X
```

```python
Scaled_X_features = Unscaled_X.columns
```

```python
Scaled_X = pd.DataFrame(columns = [Scaled_X_features], data = Scaled_X)
Scaled_X.head()
```

```python
dummy_X = df_preprocessed[['Reason for Absence_2', 'Reason for Absence_3', 'Reason for Absence_4',
       'Education_1']]
dummy_X.head()
```

```python
X_df = pd.concat([dummy_X,Scaled_X],axis = 1)
X_df.head()
```

```python
X = X_df.copy()
y = df_preprocessed['Absenteeism Time in Hours']

# Splitting the pre-processed train dataset into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify = y, random_state=42)
```

```python
# Lets check the results
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Logistic Regression Model

```python
# Hypertuning the Logistic Regression Model

parameters_grid = {'penalty':['l1', 'l2'],
                   'dual':[True, False],'C':[0.001, 0.01, 0.1, 1, 10, 100, 1000], 
                   'fit_intercept':[True, False],         
                   'solver':['saga','lbfgs','liblinear']} 

# Create grid search object with cross validating it 5 times  
LogReg_grid = GridSearchCV(LogisticRegression(random_state=42), parameters_grid, cv=5, n_jobs=2, scoring='f1_weighted') 

# Fit on data  
%time LogReg_grid.fit(X_train, y_train) 

LogReg_grid.best_params_ # printing out best parameters
```

```python
y_pred = LogReg_grid.predict(X_test)

print('Metrics of the logistic regression model after variable selection: \n')

print("Confusion matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nF1 Score = {:.5f}".format(f1_score(y_test, y_pred, average="micro")))
print("AUC Score = {:.3f}".format(roc_auc_score(y_test, y_pred)))
print("Accuracy = {:.3f}".format(accuracy_score(y_test, y_pred)))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

cf_matrix = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
(tn, fp, fn, tp)

sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True, 
            fmt='.2%', cmap='Blues')
```

# Finding Model Intercept and Coefficients

```python
# Without hyperparameter tuning
classifier_LR = LogisticRegression(random_state=42)

classifier_LR.fit(X_train, y_train)
```

```python
y_pred = classifier_LR.predict(X_test) 

print('Metrics of the logistic regression model after variable selection: \n')

print("Confusion matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nF1 Score = {:.5f}".format(f1_score(y_test, y_pred, average="micro")))
print("AUC Score = {:.3f}".format(roc_auc_score(y_test, y_pred)))
print("Accuracy = {:.3f}".format(accuracy_score(y_test, y_pred)))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

cf_matrix = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
(tn, fp, fn, tp)

sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True, 
            fmt='.2%', cmap='Blues')
```

```python
# Intercept
classifier_LR.intercept_
```

```python
# Coefficients
classifier_LR.coef_
```

```python
# Printing the coefficients in a table format
feature_names = X.columns

summary_table = pd.DataFrame(columns = ["Feature Names"], data = feature_names)
summary_table["Coefficients"] = np.transpose(classifier_LR.coef_)
summary_table
```

```python
# Adding the intercept to the summary table. 
#To add intercept to the top of the table, the index of the coefficients have to be shifted down

summary_table.index = summary_table.index + 1
summary_table.loc[0] = ["Intercept", classifier_LR.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

Note: Intercept is called bias and the coefficients are called weights. Closer to 0, smaller the weight. Farther from 0 whether positive or negative, higher is the weight. Weights can be compared this way only when the features have been normalized like we did here.

log(odds) = intercept + x1b1 + x2b2+....+xnbn. 
Changing the equation into exponents

```python
summary_table["Odds_ratio"] = np.exp(summary_table["Coefficients"])
summary_table
```

```python
# Sorting them according to the odds ratio weights
summary_table.sort_values("Odds_ratio",ascending = False)
```
